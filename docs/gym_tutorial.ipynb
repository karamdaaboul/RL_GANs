## Developing an RL cycle using OpenAI GYM
Gym is a toolkit for the development and comparison of reinforcement learning algorithms and has many built-in environments like the cart pole, pendulum.

```python
import numpy as np
import gym
from gym import spaces
from gym import ObservationWrapper

import matplotlib.pyplot as plt
```

Each environment has a register that you can use to define your environment. Under this link https://www.gymlibrary.ml/, you can find more information about your environment. What are observation space and action space? What is the reward function we are trying to maximize? 

```python
env = gym.make('HalfCheetah-v2')
```

To initialize the environment, we can use the command

```python
env.reset()
```

The interaction with a step in the environment can be done using a random action

```python
action = env.action_space.sample()
next_State, reward, done ,info = env.step(action)

print("reward: ", reward)
```

### Gym Environment Wrappers:
To extend the functionality of the environment generically. For example, if you want to be able to crop or preprocess the pixels of an image (state) to make it easier for the agent to digest, or if you want to normalize the reward points somehow. Gym offers you a convenient framework for this purpose, the so-called Wrapper class. I defined many wrappers as classes, check their code if you want.


```python
from rl_gans.wrappers.pixel_observation_wrapper import PixelObservation
```

```python
env = PixelObservation(env,observation_size= 81,normalize=False)
```
